```python
trainp=traindf.to_numpy()
testnp=testdf.to_numpy()
trainp.shape,testnp.shape
```

```python
def cost(x,y,m):
    costval = 0
    N = x.shape[0]
    for i in range(N):
        costval +=(1/N)* ((y[i]-(m* x[i]).sum())**2)
    return float(costval)
```

```python
def stepgrad(x,y,alpha,m):
    N=x.shape[0]
    updated_m=np.zeros(x.shape[1])
    for i in range(N):
        currin=x[i]
        currout=y[i]
        for j in range(x.shape[1]):
            updated_m[j] += (-2/N)*(currout-(m*currin).sum())*currin[j]
        newm=m-(alpha*updated_m)
    return newm
```

```python
def gd(x,y,alpha,num):
    m=np.zeros(x.shape[1])
    for i in range(num):
        m=stepgrad(x,y,alpha,m)
        if i % 100== 0:
            print(i, "Cost= ", cost(x,y,m))
    print("Cost=", cost(x,y,m))
    return m
```

```python
def run(trainp):
    x=trainp[:,:-1]
    x=x=np.append(x, np.ones(len(x)).reshape(-1, 1), axis=1)
    y=trainp[:,-1]
    alpha = 0.1
    num = 1500
    M = gd(x,y,alpha,num)
    return M
```

```python
m=run(trainp)
print(m)
```

```python
slopes=m
slopes
```

```python
m=slopes[0:-1]
c=slopes[-1]
m,c
```

```python
def predict(dataval, m, c):
    N = dataval.shape[0]
    y = np.zeros(N)
    for i in range(N):                               #prediction using m ,c ,alpha and numof iterations 
        X = dataval[i,:]
        y[i] = (m * X).sum() + c
    return y
```

```python
testpred=predict(testnp,m,c)
testpred,testpred.shape
```
